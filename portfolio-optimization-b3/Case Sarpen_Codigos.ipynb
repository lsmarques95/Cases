```python
# ============================
# 1) Parâmetros e identificação
# ============================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Parâmetros
window = 252  # janela móvel (1 ano útil)
acoes_cols = [col for col in df.columns if col not in ['CDI', 'IBOV']]
dias_ano = 252

print(f"Usando {len(acoes_cols)} ações. Janela = {window} dias.")
```

```python
# ====================================================
# 2) Calcular volatilidade rolling (desvio padrão diário)
#    e construir pesos inverso-volatilidade (out-of-sample)
# ====================================================

# 2.1 Volatilidade rolling (std diária) por ativo
vol_rolling = df[acoes_cols].rolling(window=window).std()

# 2.2 Para garantir OOS: shift(1) -> usamos apenas informação até t-1 para formar pesos em t
vol_for_weights = vol_rolling.shift(1)

# 2.3 Inverse-volatility weights: w_i(t) = (1 / sigma_i(t)) / sum_j (1 / sigma_j(t))
# Tratar sigmas zero/NaN: onde sigma é zero ou NaN, deixamos peso zero
inv_vol = 1.0 / vol_for_weights
inv_vol = inv_vol.replace([np.inf, -np.inf], np.nan)

# Substituir NaN por 0 para evitar problemas na normalização (esses dias serão excluídos depois)
inv_vol_filled = inv_vol.fillna(0.0)

# Normalizar por linha
weights_iv = inv_vol_filled.div(inv_vol_filled.sum(axis=1), axis=0)
# Se alguma linha tiver soma zero (por exemplo início do periodo), resultará em NaN; mantemos NaN
weights_iv[inv_vol_filled.sum(axis=1) == 0] = np.nan

# Exibir primeiras linhas para conferência
display(weights_iv.iloc[window-1:window+3].round(4))  # mostra início (após janela)
```

```python
# ====================================================
# 3) Calcular retorno diário da carteira Low-Vol (OOS)
# ====================================================

# Retorno diário da carteira low-vol (multiplicação element-wise dos pesos (no dia t) com retornos do dia t)
# Observação: weights_iv at index t uses information up to t-1 (shifted), aplicado ao retorno de t
ret_lowvol = (weights_iv * df[acoes_cols]).sum(axis=1)

# Conferir head / tail
print("Primeiros retornos Low-Vol (com NaNs até começar a janela):")
display(ret_lowvol.dropna().head())
print("Número de observações válidas (out-of-sample):", ret_lowvol.dropna().shape[0])
```

```python
# ====================================================
# 4) Definir período out-of-sample (onde ret_lowvol não é NaN)
# ====================================================
oos_mask = ret_lowvol.notna()
oos_start = ret_lowvol.index[oos_mask.argmax()]  # primeiro índice válido
oos_end = ret_lowvol.index[oos_mask.values.max() and oos_mask.values.argmax()]  # último índice com True (safer alternative below)

# safer last valid:
if oos_mask.any():
    oos_end = ret_lowvol[oos_mask].index[-1]
else:
    oos_end = None

print("Período out-of-sample começando em:", oos_start.date(), "até", oos_end.date())
```

```python
# ====================================================
# 5) MÉTRICAS — Low-Vol (OOS) vs Equal-Weighted (FULL) vs IBOV (FULL)
# ====================================================

import numpy as np
import pandas as pd

dias_ano = 252

# ----------------------------
# Função geral de métricas
# ----------------------------
def metrics_from_returns(ret_series, rf_series=None, dias_ano=252):
    s = ret_series.dropna()
    if rf_series is None:
        rf = pd.Series(0.0, index=s.index)
    else:
        rf = rf_series.reindex(s.index).ffill().fillna(0.0)

    ret_ann = s.mean() * dias_ano
    vol_ann = s.std() * np.sqrt(dias_ano)

    excesso = s - rf
    sharpe = (excesso.mean() / s.std()) * np.sqrt(dias_ano) if s.std() != 0 else np.nan

    cum = (1 + s).cumprod()
    rolling_max = cum.cummax()
    drawdown = cum / rolling_max - 1
    max_dd = drawdown.min()

    return {
        "Retorno Anual (%)": ret_ann * 100,
        "Vol (%)": vol_ann * 100,
        "Sharpe": sharpe,
        "MaxDD (%)": max_dd * 100
    }


# ================================
# 5.1 — Low-Vol OOS
# ================================
ret_lowvol_oos = ret_lowvol[oos_mask]
rf_lowvol = df["CDI"].reindex(ret_lowvol_oos.index)
ibov_lowvol = df["IBOV"].reindex(ret_lowvol_oos.index)

metrics_lowvol = metrics_from_returns(ret_lowvol_oos, rf_lowvol)

# Beta Low-Vol
cov_lv_ibov = np.cov(ret_lowvol_oos.dropna(), ibov_lowvol.dropna())[0,1]
beta_lowvol = cov_lv_ibov / ibov_lowvol.var()
metrics_lowvol["Beta"] = beta_lowvol


# ================================
# 5.2 — Equal-Weighted FULL SAMPLE (cenário-base do problema 1)
# ================================
ret_eq_full = carteira_eq                # já calculado no problema 1
rf_full = df["CDI"]

metrics_eq = metrics_from_returns(ret_eq_full, rf_full)

cov_eq_ibov = np.cov(ret_eq_full.dropna(), df["IBOV"].dropna())[0,1]
beta_eq = cov_eq_ibov / df["IBOV"].var()
metrics_eq["Beta"] = beta_eq


# ================================
# 5.3 — IBOV FULL SAMPLE
# ================================
metrics_ibov = metrics_from_returns(df["IBOV"], df["CDI"])
metrics_ibov["Beta"] = 1.0


# ================================
# 5.4 — TABELA FINAL
# ================================
resumo = pd.DataFrame({
    "Low-Vol (OOS)": metrics_lowvol,
    "Equal-Weighted (FULL)": metrics_eq,
    "Ibovespa (FULL)": metrics_ibov
}).T

# formatação
resumo_formatado = resumo.copy()
resumo_formatado["Retorno Anual (%)"] = resumo_formatado["Retorno Anual (%)"].map("{:.2f}%".format)
resumo_formatado["Vol (%)"] = resumo_formatado["Vol (%)"].map("{:.2f}%".format)
resumo_formatado["Sharpe"] = resumo_formatado["Sharpe"].map("{:.3f}".format)
resumo_formatado["MaxDD (%)"] = resumo_formatado["MaxDD (%)"].map("{:.2f}%".format)
resumo_formatado["Beta"] = resumo_formatado["Beta"].map("{:.3f}".format)

print("===== COMPARAÇÃO FINAL — LOW-VOL OOS vs CENÁRIO BASE =====")
display(resumo_formatado)
```

```python
# ====================================================
# 6) Plots do Problema 2 — Low-Vol OOS vs EW FULL vs IBOV FULL
# ====================================================

import matplotlib.pyplot as plt

# 6.1 Retorno Acumulado
# ---------------------------

# Low-Vol: somente OOS
cum_lowvol = (1 + ret_lowvol_oos).cumprod()

# EW (cenário base): FULL 2004–2019
cum_eq = (1 + carteira_eq).cumprod()

# Ibovespa: FULL 2004–2019
cum_ibov = (1 + df["IBOV"]).cumprod()

plt.figure(figsize=(12,5))
plt.plot(cum_lowvol, label='Low-Vol (OOS)', linewidth=2)
plt.plot(cum_eq, label='Equal-Weighted (FULL)', linewidth=1.5, alpha=0.80)
plt.plot(cum_ibov, label='Ibovespa (FULL)', linewidth=1.5, alpha=0.80)

plt.title("Retorno Acumulado — Low-Vol (OOS) vs EW e Ibov (FULL)")
plt.xlabel("Data")
plt.ylabel("Valor acumulado")
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.savefig("retorno_acumulado_lowvol_oos.png", dpi=300, bbox_inches='tight')
plt.show()
print("Salvo: retorno_acumulado_lowvol_oos.png")


# 6.2 Retornos anuais
# ---------------------------

# Low-Vol anual → somente OOS
ret_ano_low = ret_lowvol_oos.groupby(ret_lowvol_oos.index.year)\
                            .apply(lambda x: (1+x).prod() - 1)

# EW anual → FULL sample, mas filtramos somente anos do OOS para comparação justa
ret_ano_eq_full = carteira_eq.groupby(carteira_eq.index.year)\
                             .apply(lambda x: (1+x).prod() - 1)
ret_ano_eq = ret_ano_eq_full.reindex(ret_ano_low.index)

# IBOV anual → FULL sample filtrado para anos OOS
ret_ano_ibov_full = df["IBOV"].groupby(df.index.year)\
                               .apply(lambda x: (1+x).prod() - 1)
ret_ano_ibov = ret_ano_ibov_full.reindex(ret_ano_low.index)

anos = ret_ano_low.index

plt.figure(figsize=(14,6))
width = 0.28

plt.bar(anos - width, ret_ano_low.values, width=width, label='Low-Vol (OOS)')
plt.bar(anos, ret_ano_eq.values, width=width, label='Equal-Weighted (FULL)')
plt.bar(anos + width, ret_ano_ibov.values, width=width, label='Ibovespa (FULL)')

plt.title("Retornos Anuais — Low-Vol (OOS) vs EW e Ibov (FULL)")
plt.xlabel("Ano")
plt.ylabel("Retorno Anual")
plt.xticks(anos, rotation=45)
plt.legend()
plt.grid(axis='y')
plt.tight_layout()

plt.savefig("retornos_anuais_lowvol_vs_eq_ibov.png", dpi=300, bbox_inches='tight')
plt.show()
print("Salvo: retornos_anuais_lowvol_vs_eq_ibov.png")
```

```python
import pandas as pd
import numpy as np

# ============================================================
# Função oficial de métricas para todo o notebook
# ============================================================
def metrics(ret, rf, benchmark=None):
    ret = ret.dropna()
    rf = rf.reindex(ret.index).fillna(0)

    # Retorno anualizado
    ret_annual = ret.mean() * 252

    # Volatilidade anualizada
    vol_annual = ret.std() * np.sqrt(252)

    # Sharpe
    sharpe = ((ret - rf).mean() / ret.std()) * np.sqrt(252)

    # Max Drawdown
    cum = (1 + ret).cumprod()
    rollmax = cum.cummax()
    mdd = (cum / rollmax - 1).min()

    # Beta (se houver benchmark)
    if benchmark is not None:
        bench = benchmark.reindex(ret.index).dropna()
        aligned = ret.loc[bench.index]
        beta = np.cov(aligned, bench)[0,1] / bench.var()
    else:
        beta = np.nan

    return {
        "Retorno Anualizado (%)": ret_annual * 100,
        "Volatilidade (%)": vol_annual * 100,
        "Sharpe": sharpe,
        "Máx Drawdown (%)": mdd * 100,
        "Beta vs IBOV": beta
    }

# ============================================================
# Alinhamento das séries
# ============================================================
ret_eq_full = carteira_eq  # cenário-base do problema 1 (2004–2019)
rf_full = df['CDI']
ret_ibov_full = df['IBOV']

# Low-vol já está filtrado OOS:
ret_lowvol_oos = ret_lowvol.loc[ret_lowvol_oos.index]
rf_oos = df['CDI'].reindex(ret_lowvol_oos.index)
ibov_oos = df['IBOV'].reindex(ret_lowvol_oos.index)

# ============================================================
# Cálculo das métricas
# ============================================================
m_lowvol = metrics(ret_lowvol_oos, rf_oos, ibov_oos)
m_eq = metrics(ret_eq_full, rf_full, ret_ibov_full)
m_ibov = metrics(ret_ibov_full, rf_full, ret_ibov_full)

# ============================================================
# Comparação final
# ============================================================
comparacao = pd.DataFrame(
    [m_lowvol, m_eq, m_ibov],
    index=["Low-Vol (OOS)", "Equal-Weighted (FULL)", "Ibovespa (FULL)"]
)

display(
    comparacao.style.format({
        "Retorno Anualizado (%)": "{:.2f}",
        "Volatilidade (%)": "{:.2f}",
        "Sharpe": "{:.3f}",
        "Máx Drawdown (%)": "{:.2f}",
        "Beta vs IBOV": "{:.3f}"
    })
)
```

```python
# ===============================
# 1) TESTE T - DIFERENÇA DE MÉDIAS
# ===============================
from scipy.stats import ttest_rel
import pandas as pd

# Alinhamento
lv = ret_lowvol_oos.dropna()
eq = ret_eq_oos.reindex(lv.index).dropna()
ibov = ret_ibov_oos.reindex(lv.index).dropna()

# Teste t pareado (two-tailed)
t_eq = ttest_rel(lv, eq, nan_policy='omit')
t_ibov = ttest_rel(lv, ibov, nan_policy='omit')

print("===== TESTE T: Low-Vol vs Equal-Weighted =====")
print(t_eq)

print("\n===== TESTE T: Low-Vol vs Ibovespa =====")
print(t_ibov)

# Interpretação manual:
# p-value < 0.05 → diferença estatisticamente significativa
```

```python
# ===============================
# 2) TESTE DE SHARPE - LEDOIT & WOLF (2008)
# ===============================
import numpy as np

def ledoit_wolf_sharpe_test(r1, r2):
    """
    Teste de Ledoit-Wolf para diferença de Sharpe.
    Retorna estatística Z e p-value aproximado (two-sided).
    """
    r1 = r1.dropna()
    r2 = r2.reindex(r1.index).dropna()
    n = len(r1)
    
    # Sharpe individual
    s1 = r1.mean() / r1.std()
    s2 = r2.mean() / r2.std()
    
    # Diferença de Sharpe
    d = s1 - s2
    
    # Covariâncias necessárias
    cov11 = np.mean((r1 - r1.mean())**2)
    cov22 = np.mean((r2 - r2.mean())**2)
    cov12 = np.mean((r1 - r1.mean())*(r2 - r2.mean()))
    
    # Estimador Ledoit-Wolf
    var_d = (cov11 / (r1.std()**2) +
             cov22 / (r2.std()**2) -
             2 * cov12 / (r1.std() * r2.std())) / n
    
    z_stat = d / np.sqrt(var_d)
    
    # p-value two-sided
    from scipy.stats import norm
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    
    return d, z_stat, p_value, s1, s2

print("===== LEDOIT-WOLF: Low-Vol vs Equal-Weighted =====")
print(ledoit_wolf_sharpe_test(ret_lowvol_oos, ret_eq_oos))

print("\n===== LEDOIT-WOLF: Low-Vol vs Ibovespa =====")
print(ledoit_wolf_sharpe_test(ret_lowvol_oos, ret_ibov_oos))
```
