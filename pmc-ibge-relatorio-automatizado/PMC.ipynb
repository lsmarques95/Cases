```python
import pandas as pd
from datetime import datetime
import requests as req
import xlwings as xw
import time

start_time = time.time()

print(datetime.now())

tabela8757={'url':'https://apisidra.ibge.gov.br/values/t/8757/n1/all/v/7169,7170/p/all/c11046/56732/d/v7169%205,v7170%205'}
tabela8880={'url':'https://apisidra.ibge.gov.br/values/t/8880/n1/all/v/7169,7170/p/all/c11046/56734/d/v7169%205,v7170%205'}
tabela8881={'url':'https://apisidra.ibge.gov.br/values/t/8881/n1/all/v/7169,7170/p/all/c11046/56736/d/v7169%205,v7170%205'}
tabela8883={'url':'https://apisidra.ibge.gov.br/values/t/8883/n1/all/v/7169,7170/p/all/c11046/56736/c85/all/d/v7169%205,v7170%205'}

def pmc_8757():
  xw.App(visible=False)
  wb = xw.Book(r"C:\Users\lucas\Python\Cases\Kapitalo\Base de Dados\IBGE - SIDRA\Tabela 8757.xlsx")
  r = req.get(tabela8757.get('url'),  verify=False)
  raw = r.json()
  raw_df = pd.DataFrame(raw) 
  df_filtrado = raw_df[['V',  'D2N',  'D3C',  'D4C',  'D4N']]
  df_filtrado.columns = df_filtrado.iloc[0]
  df_filtrado.drop(0, inplace=True)
  for var in df_filtrado['Variável'].unique():
    if 'ajuste' in var:
      df = df_filtrado[df_filtrado['Variável']==var]
      date = df_filtrado['Mês (Código)'].unique()
      date = pd.to_datetime(date, format='%Y%m')
      Data = pd.DataFrame({'Data': date})
      last_date = Data['Data'].iloc[-1]
      sup_df = df[['Valor']]
      sup_df.reset_index(drop=True, inplace=True)
      dessaz_df = pd.concat([Data, sup_df], axis=1)
      wb.sheets[1]['A5'].options(index=False, header=False).value = dessaz_df
      wb.save()
    else:
      df = df_filtrado[df_filtrado['Variável']==var]
      date = df_filtrado['Mês (Código)'].unique()
      date = pd.to_datetime(date, format='%Y%m')
      Data = pd.DataFrame({'Data': date})
      last_date = Data['Data'].iloc[-1]
      sup_df = df[['Valor']]
      sup_df.reset_index(drop=True, inplace=True)
      norm_df = pd.concat([Data, sup_df], axis=1)
      wb.sheets[0]['A5'].options(index=False, header=False).value = norm_df
      wb.save()
  wb.close()
  print('Tabela 8757 atualizada, último dado:', last_date)

def pmc_8880():
  xw.App(visible=False)
  wb = xw.Book(r"C:\Users\lucas\Python\Cases\Kapitalo\Base de Dados\IBGE - SIDRA\Tabela 8880.xlsx")
  r = req.get(tabela8880.get('url'),  verify=False)
  raw = r.json()
  raw_df = pd.DataFrame(raw)
  df_filtrado = raw_df[['V',  'D2N',  'D3C',  'D4C',  'D4N']]
  df_filtrado.columns = df_filtrado.iloc[0]
  df_filtrado.drop(0, inplace=True)
  for var in df_filtrado['Variável'].unique():
    if 'ajuste' in var:
      df = df_filtrado[df_filtrado['Variável']==var]
      date = df_filtrado['Mês (Código)'].unique()
      date = pd.to_datetime(date, format='%Y%m')
      Data = pd.DataFrame({'Data': date})
      last_date = Data['Data'].iloc[-1]
      sup_df = df[['Valor']]
      sup_df.reset_index(drop=True, inplace=True)
      dessaz_df = pd.concat([Data, sup_df], axis=1)
      wb.sheets[1]['A6'].options(index=False, header=False).value = dessaz_df
      wb.save()
    else:
      df = df_filtrado[df_filtrado['Variável']==var]
      date = df_filtrado['Mês (Código)'].unique()
      date = pd.to_datetime(date, format='%Y%m')
      Data = pd.DataFrame({'Data': date})
      last_date = Data['Data'].iloc[-1]
      sup_df = df[['Valor']]
      sup_df.reset_index(drop=True, inplace=True)
      norm_df = pd.concat([Data, sup_df], axis=1)
      wb.sheets[0]['A6'].options(index=False, header=False).value = norm_df
      wb.save()
  wb.close()
  print('Tabela 8880 atualizada, último dado:', last_date)

def pmc_8881():
  xw.App(visible=False)
  wb = xw.Book(r"C:\Users\lucas\Python\Cases\Kapitalo\Base de Dados\IBGE - SIDRA\Tabela 8881.xlsx")
  r = req.get(tabela8881.get('url'),  verify=False)
  raw = r.json()
  raw_df = pd.DataFrame(raw)
  df_filtrado = raw_df[['V',  'D2N',  'D3C',  'D4C',  'D4N']]
  df_filtrado.columns = df_filtrado.iloc[0]
  df_filtrado.drop(0, inplace=True)
  for var in df_filtrado['Variável'].unique():
    if 'ajuste' in var:
      df = df_filtrado[df_filtrado['Variável']==var]
      date = df_filtrado['Mês (Código)'].unique()
      date = pd.to_datetime(date, format='%Y%m')
      Data = pd.DataFrame({'Data': date})
      last_date = Data['Data'].iloc[-1]
      sup_df = df[['Valor']]
      sup_df.reset_index(drop=True, inplace=True)
      dessaz_df = pd.concat([Data, sup_df], axis=1)
      wb.sheets[1]['A6'].options(index=False, header=False).value = dessaz_df
      wb.save()
    else:
      df = df_filtrado[df_filtrado['Variável']==var]
      date = df_filtrado['Mês (Código)'].unique()
      date = pd.to_datetime(date, format='%Y%m')
      Data = pd.DataFrame({'Data': date})
      last_date = Data['Data'].iloc[-1]
      sup_df = df[['Valor']]
      sup_df.reset_index(drop=True, inplace=True)
      norm_df = pd.concat([Data, sup_df], axis=1)
      wb.sheets[0]['A6'].options(index=False, header=False).value = norm_df
      wb.save()
  wb.close()
  print('Tabela 8881 atualizada, último dado:', last_date)

def pmc_8883():
  xw.App(visible=False)
  wb = xw.Book(r"C:\Users\lucas\Python\Cases\Kapitalo\Base de Dados\IBGE - SIDRA\Tabela 8883.xlsx")
  r = req.get(tabela8883.get('url'),  verify=False)
  raw = r.json()
  raw_df = pd.DataFrame(raw)
  df_filtrado = raw_df[['V',  'D2N',  'D3C',  'D4C',  'D5N']]
  df_filtrado.columns = df_filtrado.iloc[0]
  df_filtrado.drop(0, inplace=True)
  for var in df_filtrado['Variável'].unique():
    if 'ajuste' in var:
      df = df_filtrado[df_filtrado['Variável']==var]
      date = df_filtrado['Mês (Código)'].unique()
      date = pd.to_datetime(date, format='%Y%m')
      input_series = []
      Data = pd.DataFrame({'Data': date})
      last_date = Data['Data'].iloc[-1]
      for serie in df['Atividades'].unique():
        sup_df = df[df["Atividades"]==serie][['Valor']]
        sup_df.reset_index(inplace=True)
        input_series.append(pd.Series(sup_df['Valor'], name=serie))
      dessaz_df = pd.concat(input_series, axis=1)
      dessaz_df = pd.concat([Data, dessaz_df], axis=1)
      wb.sheets[1]['A7'].options(index=False, header=False).value = dessaz_df
      wb.save()
    else:
      df = df_filtrado[df_filtrado['Variável']==var]
      date = df_filtrado['Mês (Código)'].unique()
      date = pd.to_datetime(date, format='%Y%m')
      input_series = []
      Data = pd.DataFrame({'Data': date})
      last_date = Data['Data'].iloc[-1]
      for serie in df['Atividades'].unique():
        sup_df = df[df['Atividades']==serie][['Valor']]
        sup_df.reset_index(inplace=True)
        input_series.append(pd.Series(sup_df['Valor'], name=serie))
      norm_df = pd.concat(input_series, axis=1)
      norm_df = pd.concat([Data, norm_df], axis=1)
      wb.sheets[0]['A7'].options(index=False, header=False).value = norm_df
      wb.save()
  wb.close()
  print('Tabela 8883 atualizada, último dado:', last_date)

def pmc_atualiza():
  pmc_8757()
  pmc_8880()
  pmc_8881()
  pmc_8883()
  today = datetime.today()
  print("PMC atualizada,", today.date())
  print(time.time() - start_time)
pmc_atualiza()
print(datetime.now())
```
